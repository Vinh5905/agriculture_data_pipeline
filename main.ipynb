````python
from unstructured.partition.auto import partition

pdf_path = './data/pdf_2.pdf'

# Extract elements using Unstructured.io
unstructured_elements = partition(
    filename=pdf_path, 
    chunking_strategy="by_title",   # Tách theo heading/title nếu có thể
    languages=["vie"])

# Concatenate the text content of all extracted elements
unstructured_text = "\n".join([str(el) for el in unstructured_elements])

# Print the concatenated text and the list of extracted elements
print("--- Full text extracted using Unstructured.io ---")
print(unstructured_text[:1000] + "...") # Print first 1000 characters for brevity

print("\n--- Unstructured.io Elements (first 5) ---")
for i, element in enumerate(unstructured_elements[:5]):
    print(f"Element Type: {type(element)}")
    print(f"Text: {element}\n")
````

````python
import re

# Define the regex pattern for identifying titles
# It looks for lines starting with:
# - One or more Roman numerals followed by a dot and space (e.g., I., II.)
# - One or more groups of digits followed by a dot, ending with a dot and space (e.g., 1., 1.1., 2.3.4.)
# - A lowercase letter followed by a closing parenthesis or a dot, and a space (e.g., a), b., c.)
# - The keywords "BÀI", "CHƯƠNG", "PHẦN" followed by a space or colon and one or more digits (case-insensitive)
# The .*$ captures the rest of the line.
title_pattern = re.compile(r"^\s*([IVXLCDM]+\.\s+|(\d+\.)+\s+|[a-z][).]\s+|(bài|chương|phần)\s+\d+[:\s]+).*$", re.MULTILINE | re.IGNORECASE)

print("Regex pattern defined.")
````

````python
titles_with_positions = []
full_text_with_blocks = ""
current_char_index = 0

# Iterate through the elements to find titles and record their positions
for element in unstructured_elements:
    element_text = str(element)
    print("Text: ", element_text)
    # Find all matches of the title pattern within the element'text
    for match in title_pattern.finditer(element_text):
        start, end = match.span()
        # Calculate the absolute start and end character indices in the full text
        absolute_start = current_char_index + start
        absolute_end = current_char_index + end
        titles_with_positions.append({
            "title": match.group(0).strip(),
            "start_char_index": absolute_start,
            "end_char_index": absolute_end
        })

    full_text_with_blocks += element_text + "\n" # Build the full text with blocks separated by newline
    current_char_index += len(element_text) + 1 # Update the current character index, accounting for the added newline

print("--- Titles with Positions (first 5) ---")
for i, title_info in enumerate(titles_with_positions[:5]):
    print(f"Title: {title_info['title']}")
    print(f"Start Index: {title_info['start_char_index']}")
    print(f"End Index: {title_info['end_char_index']}")
    print("-" * 20)

# The 'titles_with_positions' list now contains the extracted titles and their character positions.
# The 'full_text_with_blocks' string contains the concatenated text of all elements.
````

````python
for i in titles_with_positions:
  print(i['title'])
````

````python
# Calculate the end character index for each title segment
for i in range(len(titles_with_positions)):
    if i < len(titles_with_positions) - 1:
        # The end of the current segment is the start of the next segment
        titles_with_positions[i]['segment_end_char_index'] = titles_with_positions[i+1]['start_char_index']
    else:
        # The end of the last segment is the end of the full text
        titles_with_positions[i]['segment_end_char_index'] = len(full_text_with_blocks)

print("--- Updated Titles with Segment End Positions (first 5) ---")
for i, title_info in enumerate(titles_with_positions):
    print(f"Title: {title_info['title']}")
    print(f"Start Index: {title_info['start_char_index']}")
    print(f"End Index: {title_info['end_char_index']}")
    print(f"Segment End Index: {title_info['segment_end_char_index']}")
    print("-" * 20)
````

````python
# Chuyển đổi danh sách các từ điển thành danh sách các danh sách
converted_titles_with_positions = []
for item in titles_with_positions:
    converted_titles_with_positions.append([
        item['title'],
        item['start_char_index'],
        item['end_char_index'],
        item['segment_end_char_index']
    ])

# In ra kết quả để kiểm tra
for row in converted_titles_with_positions:
    print(row)
````
